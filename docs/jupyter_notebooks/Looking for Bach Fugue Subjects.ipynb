We use the algorithms through the "Finder" class. Let's start by importing the Finder responsible for using the University of Helsinki's P, S, and W algorithms:

```python
from geometric_helsinki import Finder
```

Finder objects are generators. They receive as input a 'pattern' and 'source'. Then on each iteration, a Finder will return an occurrence of the pattern within the source, until it has exhaustively looked through the entire source and yields a StopIteration exception. 

We'll start by making our first Finder object. The pattern and source inputs can be paths to music files, or music21 Stream objects.

```python
pattern = 'music_files/bach_wtc1/e_wtc1f02_S.xml'
source = 'music_files/bach_wtc1/wtc1f02.krn'
foo = Finder(pattern, source)
```

Let's find the first statement of the subject in BWV 847:

```python
bar = next(foo)
bar
```

We've found the first occurrence! These are excerpts of the 'source' score, stored stored in a music21.stream.Score object. 
The Finder will generate a list of "matching pairs" between the pattern notes and the source notes. Each pattern note that we're looking for has a matching source note in the source. In the excerpt, the matching source note is coloured red.

We can use music21.stream.show() to print an .xml file and see it in your default MusicXML reader. Or you can use .show('lily.pdf') to render it with Lilypond and use any pdf viewer.

```python
# This will open the temp.xml file with your default MusicXML software
bar.show() 

# This will open the rendered score with your default image viewer
# bar.show('lily.pdf')
```

What about the next occurrence?

```python
next(foo).show()
```

Alright well that's cool. Be careful to remember that 'foo' is a generator. The occurrence shown above is no longer in memory; it exists only in the printed xml score. If we didn't use .show(), it'd be gone to the ether forever! 

At least until we run update(). This will restart foo at the beginning again. Now, what about the counter-subject? Can we look for that too?

```python
foo.update(pattern='music_files/bach_wtc1/e_wtc1f02_CS.xml')
for occ in foo:
    occ.show()
```

The update() function can change any of the Finder settings. 

```python
foo.update(colour='green', scale='warped')
for occ in foo:
    occ.show()
```
